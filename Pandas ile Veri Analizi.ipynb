```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar grafiği ayarları: x ekseninde cinsiyet, y ekseninde hayatta kalma oranı gösterilir
sns.barplot(x='Sex', y='Survived', data=df, ci=None, palette=["pink", "blue"])

# Çubukların üstüne yüzdelik oran eklemek için
for container in plt.gca().containers:
    # Çubuğun yüksekliğini yüzdelik forma çevirip etiket oluşturur
    labels = [f'{int(value.get_height() * 100)}%' for value in container]
    # Çubukların üstüne yüzdelik etiketleri yerleştirir
    plt.bar_label(container, labels=labels)

# Grafik etiketleri ve başlığı
plt.xlabel("Cinsiyet")  # X ekseni etiketi
plt.ylabel("Hayatta kalma oranı")  # Y ekseni etiketi
plt.title("Cinsiyete göre hayatta kalma")  # Grafik başlığı

# Grafik gösterimi
plt.show()
```
